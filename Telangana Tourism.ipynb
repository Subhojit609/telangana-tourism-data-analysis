## CAGR

```python
# Group the dataframe by 'district'
grouped = year_wise_visitors.groupby('district')

# Calculate CAGR for each district
cagr_df = pd.DataFrame(columns=['district', 'CAGR'])

for district, data in grouped:
    beginning_value = data.iloc[0]['total visitors']
    ending_value = data.iloc[-1]['total visitors']
    num_years = len(data)
    
    cagr = (ending_value / beginning_value) ** (1 / num_years) - 1
    
    cagr_df = cagr_df.append({'district': district, 'CAGR': cagr}, ignore_index=True)

# Display the dataframe with CAGR values
cagr_df.sort_values(by=['CAGR'], ascending= False, inplace=True)
cagr_df.reset_index(drop=True, inplace=True)
cagr_df
```

### Top 3 districts with CAGR

```python
highest_cagr= cagr_df[:3]
lowest_cagr= cagr_df[-3:]
```

```python
# Set a custom color palette
colors = sns.color_palette('Spectral')
sns.set_theme(style="whitegrid", palette='Spectral')

# Set the figure size
plt.figure(figsize=(20, 9))
plt.suptitle("District Wise CAGR", fontsize=28)

# Create the first subplot for df
plt.subplot(1, 2, 1)
sns.barplot(x='district', y='CAGR', data=highest_cagr, palette=colors)

plt.xlabel('District', fontsize=24, fontweight='bold')
plt.ylabel('CAGR', fontsize=24, fontweight='bold')
plt.title('Highest', fontsize=24, fontweight='bold')


sns.despine()

plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.subplot(1, 2, 2)
sns.barplot(x='district', y='CAGR', data=lowest_cagr, palette=colors)


plt.xlabel('District', fontsize=22, fontweight='bold')
plt.ylabel('')
plt.title('Lowest', fontsize=22, fontweight='bold')

# Set the y-label formatting as whole numbers
plt.ticklabel_format(style='plain', axis='y')


sns.despine()

# Set the tick sizes for dfx
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

# Adjust the plot layout
plt.tight_layout()


# Show the plot
plt.show()
```
